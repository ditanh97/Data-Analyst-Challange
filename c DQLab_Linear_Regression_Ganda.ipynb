# Definisi
Analisis regresi digunakan untuk mengukur seberapa besar pengaruh antara variabel bebas dan variabel terikat. Regresi linear berganda merupakan model regresi yang melibatkan lebih dari satu variabel independen. Analisis regresi linear berganda dilakukan untuk mengetahui arah dan seberapa besar pengaruh variabel independen terhadap variabel dependen.


# Bentuk Umum

Bentuk umum persamaan regresi linear ganda adalah sebagai berikut
\begin{equation}
Y=b_0 + b_1 X_1 + b_2 X_2
\end{equation}

dengan 
𝑌 = variabel independen
𝑋 = variabel dependen
𝑏 = koefisien regresi

# Tugas

Buatlah sebuah fungsi bernama regresi_linear_ganda dengan parameter dataX dan dataY atau <b>regresi_linear_ganda</b>(dataX, dataY) dengan keluaran berupa pasangan b0, b1, b2 dengan tipe data dict. 

Contoh :

    >>> dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
    >>> dataY = [6, 7, 8, 8, 9, 10, 5]
    >>> b = regresi_linear_ganda(dataX, dataY)
    >>> print(b)
    ... {b0 : 2.098049, b1 : -0.0154570, b2 : 0.4390340}


```python
dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
dataY = [6, 7, 8, 8, 9, 10, 5]
```

```python
def avg(arr):
    return sum(arr)/len(arr)
def mul_sum(arr):
    res = 1
    for a in arr:
        res *= a
    return res
        


def regresi_linear_ganda(dataX, dataY):
    Y_avg = avg(dataY)
    sum_Y = sum(dataY)
    X_arr = {
        'X': [],
        'sum_X': [],
        'X_avg': [],
        'x': [],
        'sum_xy': []
    }
    n = 0

    
    for i in dataX:
        Xi = dataX[i]
        sum_Xi=  sum(Xi)
        Xi_avg = avg(Xi)
        n = len(Xi)
        Xi_2_sum = sum([x**2 for x in Xi])
        xi = Xi_2_sum - (sum_Xi**2/n)
        sum_XiY = 0
        sum_xiY = 0
        for x,y in zip(Xi, dataY):
            sum_XiY += x*y 
            sum_xiY += (x*y)**2
        sum_xiy = sum_XiY - (sum_Xi*sum_Y/n)
        X_arr['X'].append(Xi)
        X_arr['sum_X'].append(sum_Xi)
        X_arr['X_avg'].append(Xi_avg)
        X_arr['x'].append(xi)
        X_arr['sum_xy'].append(sum_xiy)
    
    sumX_mul = 0
    for unit in zip(*X_arr['X']):
        mul = mul_sum(unit)
        sumX_mul += mul

    
    
    sumx_mul = sumX_mul - mul_sum(X_arr['sum_X'])/n
    
    print(sumx_mul)
    
    b1 = (X_arr['x'][1]*X_arr['sum_xy'][0] -  sumx_mul*X_arr['sum_xy'][1])/(X_arr['x'][0]*X_arr['x'][1]- sumx_mul**2)
    
    b2 = (X_arr['x'][0]*X_arr['sum_xy'][1] - sumx_mul*X_arr['sum_xy'][0])/(X_arr['x'][0]*X_arr['x'][1]- sumx_mul**2)
    
    b0 = Y_avg - b1*X_arr['X_avg'][0]-b2*X_arr['X_avg'][1]
    
    res = {'b0': b0, 'b1': b1, 'b2': b2 }

    return res

b = regresi_linear_ganda(dataX, dataY)
print(b)
```
