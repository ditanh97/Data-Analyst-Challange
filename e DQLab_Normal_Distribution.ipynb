# Definisi

Distribusi normal merupakan fungsi probabilitas yang menunjukkan adanya distribusi atau penyebaran suatu variabel. Fungsi dari distribusi normal tersebut biasanya dibuktikan oleh adanya grafik simetris yang disebut kurva lonceng atau bell curve yang menandakan adanya distribusi yang merata, sehingga kurva akan memuncak di bagian tengah dan melandai di kedua sisi dengan nilai yang setara.

Variabel acak kontinu adalah variabel acak yang mempunyai nilai-nilai tak terhingga dan tak terbilang. Dimana jika X adalah variabel acak dan P(ùëã = ùòπ) adalah distribusi probabilitas dari ùëã, maka fungsi f(x) = P(ùëã = ùòπ) disebut fungsi kepadatan peluang (<i>probability density function</i>). 

Pada distribusi normal fungsi kepadatan peluang dinyatakan sebagai berikut :

\begin{equation}
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\:e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}},\;\;\forall\;x\:\in\:(-\infty\:,\:\infty)
\end{equation}

<b>Contoh:</b>

Sebuah data berat badan mahasiswa perguruan tinggi X berdistribusi normal dengan rataan ùúá = 70 (kg) dengan simpangan baku œÉ = 3.5 (kg). Maka grafik sebaran peluangnya disajikan sebagai berikut

Misalkan ùëã peubah acak kontinu yang memiliki fungsi kepadatan peluang ùëì(ùë•).
Fungsi distribusi kumulatif (<i>cumulative distribution function</i> / cdf) untuk peubah acak kontinu ùëã didefinisikan sbb:

\begin{equation}
F(x)=P(X<=x) = \int_{-\infty}^{x} f(x)\;dx, \;\;\forall\;x\:\in\:(-\infty\:,\:\infty)
\end{equation}

Dengan demikian <i>cumulative distribution function</i> dari distribusi normal adalah

\begin{equation}
F(x)=P(X<=x) = \int_{-\infty}^{x}\frac{1}{\sqrt{2\pi\sigma^2}}\:e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}}\;dx, \;\;\forall\;x\:\in\:(-\infty\:,\:\infty)
\end{equation}

Note \:

  \begin{equation}
  \int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi\sigma^2}}\:e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}}\;dx = 1
  \end{equation}

# Integral Riemann

Beberapa integral dengan fungsi tertentu sangat sulit untuk dikerjakan secara eksak (hasil presisi pasti). Untuk itu digunakan pendekatan lain dalam menyelesaikan persoalan integral tertentu yakni dengan pendekatan Riemann. Pendekatan Riemann adalah salah satu aplikasi yang sangat umum untuk mengaproksimasi area dibawah grafik fungsi atau garis pada grafik, tetapi juga panjang kurva dan aproksimasi lainnya. Ide dasar integral Riemann adalah sebagai berikut

Diberikan sebuah fungsi ùëì : \[ ùëé , ùëè \] ‚Üí ‚Ñú yang terdefinisi pada interval tertutup \[ ùëé , ùëè \] dari bilangan real. Diberikan pula 
\begin{equation}
P = (x_0, x_1, x_2, ..., x_n)
\end{equation}
sebuah partisi dari \[ ùëé , ùëè \] yang mana 
\begin{equation}
a = x_0 < x_1 < x_2 ... < x_n = b
\end{equation}
Sebuah jumlahan Riemann pada fungsi ùëì dengan partisi ùëÉ didefinsikan sebagai 
\begin{equation}
\int_{a}^{b}f(x)\;dx = \lim_{n \to \infty}\sum_{i=0}^{n-1}f(x_i)\Delta x
\end{equation}

Dimana
\begin{equation}
\Delta x = \frac{b - a}{h}
\end{equation}

dengan ùíâ = banyak partisi yang diinginkan 


Integral Riemann terbagi dalam beberapa cara pendekatan :     
<ol>
  <li>Riemann Kiri</li>
  <li>Riemann Tengah</li>
  <li>Riemann Kanan</li>
</ol>

\begin{equation}
f(x)=\frac{1}{\sqrt{2\pi(50)^2}}\:e^{-\frac{1}{2}\left(\frac{x-800}{50}\right)^{2}},\;\;\forall\;x\:\in\:(-\infty\:,\:\infty)
\end{equation}

dengan nilai dari konstanta œÄ = 3.14 dan ùëí = 2.71

### Ditanya :     
  1. Peluang ditemukan karyawan dengan gaji diantara <span>&#36;750</span> sampai dengan <span>&#36;900</span> atau 

\begin{equation}
F(x)=P(750 <= x <= 900)
\end{equation}

Jawab :   
\begin{equation}
P(750 <= x <= 900) = \int_{750}^{900}f(x)\;dx
\end{equation}
\begin{equation}
P(750 <= x <= 900) = \int_{750}^{900}\frac{1}{\sqrt{2\pi(50)^2}}\:e^{-\frac{1}{2}\left(\frac{x-800}{50}\right)^{2}}\;dx
\end{equation}

Untuk menghitung integral tersebut digunakan pendekatan Riemann Tengah:      


1.   Tentukan banyak partisi (ùëÅ)

2.   Tentukan lebar tiap pias 
\begin{equation}
\Delta x = \frac{b-a}{h} = \frac{900-750}{5} = 30
\end{equation}
3.   Tentukan metode integral Riemann
     Terdapat tiga tipe perhitungan integral Riemann, yakni riemann kiri, tengah ataupun kanan. Ketiga tipe ini hanya berbeda pada penentuan ùë• nya saja.
     <ol>
     <li>Riemann Kiri</li>
        Jika digunakan metode Riemann kiri, maka ùë• dimulai dari batas bawah (pada kasus ini 750) lalu ditambah dengan Œîùë• dan seterusnya sehingga menjadi (ùë• = 750, 780, 810, 840, 870)
      <li>Riemann Kanan</li>
        Jika digunakan metode Riemann kanan, maka ùë• dimulai dari batas bawah langsung ditambah dengan Œîùë• (pada kasus ni menjadi 750 + 30 = 780) dan seterusnya sehingga menjadi (ùë• = 780, 810, 840, 870, 900)
      <li>Riemann Tengah</li>
        Jika digunakan metode Riemann tengah, maka ùë• dimulai dari batas bawah langsung ditambah dengan setengah Œîùë• pada awal perhitungan (pada kasus ni menjadi 750 + 30/2 = 765) untuk perhitungan selanjutnya tambahkan dengan Œîùë• dan seterusnya sehingga menjadi (ùë• = 765, 795, 825, 855, 885)
     </ol>

```python
a = float(input('Masukan batas atas (a) '))
```

```python
jumlah_partisi = int(input('Masukan jumlah partisi '))
```

```python
def distribusi_normal(mean : float, stdev: float, batas, jumlah_partisi: int) -> float:
    print(mean, stdev, batas, jumlah_partisi)
    x_low,x_up = batas
    del_x = (float(x_up) - float(x_low)) / (jumlah_partisi)
    result = 0
    for idx_pias in range(jumlah_partisi):
        x_pias = x_low + del_x/2 
        e_term = ((x_pias-mean)/stdev)**2
        e = math.exp(-1/2*e_term)
        k = 1/(math.sqrt(math.pi*2*(stdev)**2))
        p_pias = e*k
        result += p_pias*del_x
        #update x_low with riemann kiri
        x_low += del_x
    return result
```

```python
result = distribusi_normal(mean, stdev, (a, b), jumlah_partisi)
print(result)
```

```python
result = distribusi_normal(mean, stdev, (a, b), 1000)
print(result)
```

```python
result = distribusi_normal(mean, stdev, (a, b), 5000)
print(result)
```
